```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"D:\21bad014_ML\Rice_Sample.csv")
df
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
arr=ms.fit_transform(df.iloc[:,:-1])
arr
```

```python
df1=pd.DataFrame(arr,columns=df.columns[:-1])
df1['CLASS']=df['CLASS']
df1
```

```python
from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
df1['CLASS']=lc.fit_transform(df1['CLASS'])
df1
```

```python
scaled_data=df1.iloc[:,:-1]
```

```python
principal=PCA(n_components=3)
principal.fit(scaled_data)
x=principal.transform(scaled_data)
print(x.shape)
```

```python
principal.components_
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.scatter(x[:,0],x[:,1],c=df1['CLASS'],cmap='plasma')
plt.xlabel('pc1')
plt.ylabel('pc2')
```

```python
fig = plt.figure(figsize=(14,9))
ax = fig.add_subplot(projection='3d')
ax.scatter(x[:,0], x[:,1], x[:,2], c=df1['CLASS'])
plt.show()
```

```python
print(principal.explained_variance_ratio_)
```
