```python
def initialize_centroids(data,num):
    cent=[]
    while(len(cent)!=num):
        random_cent=random.choice(data)
        if np.all(np.array([np.all(row != random_cent) for row in cent])):
            cent.append(random_cent)
    return cent

def kmeans(data,centroids):
    print()
    print("****")
    old_centroids=centroids
    new_centroids=[]
    label=[]
    for i in range(len(old_centroids)):
        label.append([])
    for i in data:
        lis=[]
        for cent in old_centroids:
            temp=np.asarray(i)-np.asarray(cent)
            eucl_dist=np.sqrt(np.dot(temp.T,temp))
            lis.append(eucl_dist)
        print(lis)
        clus=np.argmin(lis)
        print(clus)
        label[clus].append(i)
    print(label)
    print(len(label))
    
    #Calculating the centroids
    for i in label:
        new_centroids.append(np.mean(np.array(i),axis=0))
    print(new_centroids)
    
    if(np.array_equal(new_centroids,old_centroids)):
        return new_centroids
    else:
        kmeans(data,new_centroids)
    
```
