```python
import numpy as np
def activation(net_inp,ch):
    if(ch==1):
        return net_inp
    elif(ch==2):
        return 1 if net_inp>=thresh else 0
    elif(ch==3):
        return 1 if net_inp>=thresh else -1
    elif(ch==4):
        return round(1/(1+np.exp(-net_inp)))
    elif(ch==5):
        return round((1-np.exp(-net_inp))/(1+np.exp(-net_inp)))
    
    elif(ch==6):    #RAMP FUNCTION
        if(net_inp>1):
            return 1
        elif(net_inp<0):
            return 0
        else:
            return net_inp

def predict(inputs,weights,bias,ch):
    inp=np.dot(inputs,weights)+bias
    print("--->",inputs)
    print("****",np.dot(inputs,weights))
    return activation(inp,ch)

def feed_forward(data,target,weights,bias,lr,ch):
    converged=False
    epoch=0
    while not converged and epoch<=10:
        converged=True
        print("EPOCH : ",epoch)
        print("Weights : ",weights)
        print("Bias :",bias)
        for i in range(len(data)):
            pred=predict(data[i],weights,bias,ch)
            error=target[i]-pred
            print("for data ",data[i],"PRED is ",pred)
            if error!=0:
                print("Updating Weights : ")
                weights+=lr*error*data[i]
                bias+=lr*error
                print("New Weights : ",weights)
                print("New Bias :",bias)
                converged=False
        epoch+=1
    print('\nFinal weights:',weights)
    print('Bias:',bias,"\n")
    return weights,bias


X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

test_data=np.array([[0, 0],[0, 1],[1, 0],[1, 1]])
print('AND GATE\n')
weights=[]
for i in range(len(X[0])):
    val=float(input("Enter the weight{}".format(i)))
    weights.append(val)
    
weights=np.array(weights)
print(weights)
bias=float(input("Enter the bias:"))
lr=float(input("Enter the learning rate:"))
ch=int(input('\n1.Identity Function\n2.Binary Step Fuction\n3.Bipolar Step Function\n4.Binary Sigmoidal Function\n5.Bipolar Sigmoidal Function\n6.Ramp Function\n\nEnter your choice:'))

if(ch==3 or ch==5):
    for i in range(len(y)):
        if(y[i]==0):
            y[i]=-1
            
if(ch==2 or ch==3):
    thresh=float(input('Enter the threshold:'))

weights,bias=feed_forward(X,y,weights,bias,lr,ch)

for test in test_data:
    pred=predict(test,weights,bias,ch)
    print("Input:",test,"Prediction:",pred)
```
