```python
df.columns=df.columns.str.strip()
df['Classes']=df['Classes'].str.strip()
df['Classes'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Classes']=le.fit_transform(df['Classes'])
df
```

```python
df=df.drop('Region',axis=1)
```

```python
df=df.drop('year',axis=1)
```

```python

def activation(net_inp,ch):
    if(ch==1):
        return net_inp
    elif(ch==2):
        return round(1/(1+np.exp(-net_inp)))
    elif(ch==3):
        return round((1-np.exp(-net_inp))/(1+np.exp(-net_inp)))

def predict(inputs,weights,bias,ch):
    inputs=np.dot(inputs,weights)+bias
    return activation(inputs,ch)

def feed_forward(data,target,weights,bias,lr,ch):
    converged=False
    epoch=0
    print("\len ",len(data))
    while not converged and epoch<=10:
        converged=True
        print("EPOCH : ",epoch)
        print("Weights : ",weights)
        print("Bias :",bias)
        print(data[0])
        for i in range(len(data)):
            pred=predict(data[i],weights,bias,ch)
            error=target[i]-pred
            #print("for data ",data[i],"PRED is ",pred)
            if error!=0:
             #  print("Updating Weights : ")
                weights+=lr*error*data[i]
                bias+=lr*error
              # print("New Weights : ",weights)
               #print("New Bias :",bias)
                converged=False
        epoch+=1
   # print('\nFinal weights:',weights)
    #print('Bias:',bias,"\n")
    return weights,bias
```

```python
X=np.array(df.iloc[:,:-1])
y=np.array(df['Classes'])

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_train.shape
```

```python
weights=np.random.randn(len(df.columns[:-1]))
print(weights)
weights.shape
```

```python
bias=float(input("Enter the bias:"))
lr=float(input("Enter the learning rate:"))
ch=int(input("1.IDENTITY FXN\n2.BINARY SIGMOIDAL\n3.BIPOLAR SIGMOIDAL\nEnter the Choice of A.F : "))
if(ch==2 or ch==3):
    thresh=float(input('Enter the threshold:'))

weights,bias=feed_forward(X_train,y_train,weights,bias,lr,ch)
print("\nFinal Weights and Bias : ",weights,bias)

pred=[]
for test in X_test:
    p=predict(test,weights,bias,ch)
    pred.append(p)
    print("Input:",test,"Prediction:",p)
print(pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred))
```
