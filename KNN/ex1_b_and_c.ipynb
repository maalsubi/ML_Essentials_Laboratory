```python
import pandas as pd
```

```python
df1=pd.read_csv(r'D:\21bad014_ML\new_forest_full_dataset_with_missing.csv')
df1
```

```python
df1.columns=df1.columns.str.strip()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1['Classes']=df1['Classes'].str.strip()
df1['Classes'].value_counts()
```

```python
# Handling missing values
mean=df1['Temperature'].mean()
print(mean)
df1['Temperature']=df1['Temperature'].fillna(mean)
#replace Ws WITH MEDIAN
median=df1['Ws'].median()
print(median)
df1['Ws']=df1['Ws'].fillna(median)

mean=df1['Rain'].mean()
print(mean)
df1['Rain']=df1['Rain'].fillna(mean)

#replace DMC WITH MEDIAN
median=df1['DMC'].median()
print(median)
df1['DMC']=df1['DMC'].fillna(median)

#replace DC with Mode
mode=df1['DC'].mode()[0]
print(mode)
df1['DC']=df1['DC'].fillna(mode)

#replace ISI with mean
mean=df1['ISI'].mean()
print(mean)
df1['ISI']=df1['ISI'].fillna(mean)
```

```python
#Changing the category region to numeric 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['Region']=le.fit_transform(df1['Region'])
reg_map = dict(zip(le.classes_, range(len(le.classes_))))
print(reg_map)
```

```python

from sklearn.preprocessing import StandardScaler
dfnew=df1.copy()
sc=StandardScaler()
sc.fit_transform(dfnew.iloc[:,:-1])
```

```python
#KNN SKLEARN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X=df1.iloc[:,:-1]
y=df1['Classes']
X,y
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print(X_train.shape,X_test.shape)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
train_acc=[]
test_acc=[]

kmax=int(input("Enter the max k value"))
kvalues=[i for i in range(1,kmax+1)]
for i in range(1,kmax+1):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred=knn.predict(X_test) # test_prediction
    test_acc.append(accuracy_score(y_test,pred))
    
    #train _prediction
    train_pred=knn.predict(X_train)
    train_acc.append(accuracy_score(y_train,train_pred))
print(train_acc)
print(test_acc)
```

```python
import matplotlib.pyplot as plt
plt.plot(kvalues,train_acc)
plt.xlabel('K values')
plt.ylabel('Train accuracy')
```
